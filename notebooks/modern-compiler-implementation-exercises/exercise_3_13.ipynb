# Exercise 3.13

Show that this grammar is LALR(1) but not SLR

```python
from parsers.lr.lr1 import LR1Parser
from parsers.lr.slr import SLRParser
```

```python
grammar = [
    "S -> X $",
    "X -> M a",
    "X -> b M c",
    "X -> d c",
    "X -> b d a",
    "M -> d",
]
```

```python
lalr1 = LR1Parser(*grammar).to_lalr()
```

```python
print(lalr1.to_tabulate())
```

```python
slr = SLRParser(*grammar)
```

```python
print(slr.to_tabulate())
```

There are shift/reduce conflicts in SLR parsing table, which didn't happen for LALR(1)
table.
