## Exercise 3.11

Construct LR(0) parser for given grammar and then determine whether it is
SLR grammar or not.

```python
from parsers.lr.lr0 import LR0Parser
from parsers.lr.slr import SLRParser
```

```python
grammar = [
    "S -> B $",
    "P ->",
    "P -> ( E )",
    "B -> id P",
    "B -> id ( E ]",
    "E -> B",
    "E -> B , E",
]
parser = LR0Parser(*grammar)
```

```python
parser.print_rules_and_states()
```

```python
print(parser.to_tabulate())
```

I think this grammar could be SLR one - the shift/reduce conflicts are in places
where parser SLR would insert reduce rules based on the follow items. I will
still check it building the SLR parsing table

```python
slr = SLRParser(*grammar)
```

```python
print(slr.to_tabulate())
```

As expected - no conflict in SLR table
