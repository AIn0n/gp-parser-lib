# LR(1) -> LALR(1) conversion

Let's see how the conversion works

```python
from IPython.core.display import SVG

from parsers.lr.lr1 import LR1Parser
from parsers.lr.to_graphviz import lr_parser_to_graphviz
from parsers.grammars.example_grammars import GRAMMAR_3_26
from parsers.grammars.appel_grammar_parser import appel_to_ruleset
```

```python
p = LR1Parser(appel_to_ruleset(GRAMMAR_3_26))
dot = lr_parser_to_graphviz(p, title="grammar 3.26 as LR(1)")
SVG(dot.create_svg())
```

```python
print(p.to_tabulate())
```

```python
p.print_rules_and_states()
```

```python
lalr1 = p.to_lalr(verbose=True)
```

```python
SVG(lr_parser_to_graphviz(lalr1).create_svg())
```
