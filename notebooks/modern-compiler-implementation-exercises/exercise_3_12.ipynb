# Exercise 3.12

Build LR(0) DFA for given grammar.

```python
from IPython.core.display import SVG

from parsers.lr.lr0 import LR0Parser
from parsers.lr.lr1 import LR1Parser
from parsers.lr.slr import SLRParser
from parsers.lr.to_graphviz import lr_parser_to_graphviz
```

```python
grammar = [
    "S -> E $",
    "E -> id",
    "E -> id ( E )",
    "E -> E + id",
]
```

```python
lr0 = LR0Parser(*grammar)
```

```python
SVG(lr_parser_to_graphviz(lr0).create_svg())
```

## A

is this LR(0) grammar? Give evidence

```python
print(lr0.to_tabulate())
```

No, it's not LR(0) because there is conflict in LR(0) generated parsing table.

## B

Is this an SLR grammar? Give evidence.

```python
slr = SLRParser(*grammar)
print(slr.to_tabulate())
```

Yes, this is SLR compatible grammar, because there is no conflict in parsing table.

## C

Is this LR(1) grammar? Give evidence.

```python
lr1 = LR1Parser(*grammar)
print(lr1.to_tabulate())
```

Also no conflict, so it can be LR(1) grammar.

## Is this LALR(1)?

I'm just curious if that's the case here

```python
lalr1 = lr1.to_lalr()
print(lalr1.to_tabulate())
```

Yes, it is indeed!
